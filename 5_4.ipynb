```python
var = input('   : ').split()
func = input(' : ')
c_limit = int(input('   : '))
limit = []
for i in range(c_limit):
    limit.append(input(' : '))
```

```python
def CoefficientIntercept(equation, var):
    coef = {}
    for i in range(len(var)):
        coef[var[i]] = float(re.findall(f'-?[0-9.]*[{var[i]}]', equation)[0][:-1])
    
    intercept = re.sub(f"[+-]?\d+[a-zA-Z]|[+-]?\d+\.\d+[a-zA-Z]","", equation)    

    return coef, intercept

```

```python
coef_func = CoefficientIntercept(func, var)

coef_lim = []

for i in range(len(limit)):
    coef_lim.append(CoefficientIntercept(limit[i], var))


obj = []
for i in var:
    obj.append(float(coef_func[0][i]))



lhs_eq = []
rhs_eq = []
for i in range(len(limit)):
    lhs_eq.append(list(coef_lim[i][0].values()))
    rhs_eq.append(float(coef_lim[i][1][1:]))
```

```python
opt = linprog(c=obj, A_eq=lhs_eq, b_eq=rhs_eq,
              method="simplex")

print(f'  : {opt.x}')
print(f' : {opt.fun}')
```

```python
def f1(x):
    return 4.0*x-2.0
def f2(x):
    return 0.5*x+2.0
def f3(x):
    return -0.3*x+7.0

x = Symbol('x')
x1, =  solve(f1(x)-f2(x))
x2, =  solve(f1(x)-f3(x))
x3, =  solve(f2(x)-f3(x))

y1 = f1(x1)
y2 = f1(x2)
y3 = f2(x3)

plt.plot(x1,f1(x1),'go',markersize=10)
plt.plot(x2,f1(x2),'go',markersize=10)
plt.plot(x3,f2(x3),'go',markersize=10)

plt.fill([x1,x2,x3,x1],[y1,y2,y3,y1],'red',alpha=0.5)

xr = np.linspace(0.5,7.5,100)
y1r = f1(xr)
y2r = f2(xr)
y3r = f3(xr)

plt.plot(xr,y1r,'k--')
plt.plot(xr,y2r,'k--')
plt.plot(xr,y3r,'k--')

plt.xlim(0.5,7)
plt.ylim(2,8)

plt.show()
```
